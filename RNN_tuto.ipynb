```python
# importer les packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense 
from tensorflow.keras.layers import LSTM
```

```python
# Ouvrir la base de donne√©e
df = pd.read_csv('/content/sample_data/AirPassengers.csv')
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
data = df[['#Passengers']].values
data
```

```python
# Pr√©paration de la base de donn√©es
sc = MinMaxScaler(feature_range=(0,1))
data_scaled = sc.fit_transform(data)
data_scaled
```

```python
x_data= []
y_data= []
look_back = 1
for i in range(len(data_scaled)-1):
  x_data.append(data_scaled[i:(i+look_back),0])
  y_data.append(data_scaled[i+look_back, 0])
x_data, y_data = np.array(x_data),np.array(y_data) 
print(x_data)
print(y_data)
```

```python
# Reshaping / creer trois dimension de notre x_data
x_data = np.reshape(x_data,(x_data.shape[0], x_data.shape[1],1))
```

```python
x_data.shape
```

```python
# S√©parer mes donn√©es de Test and Train 
train_size = int(len(x_data)* 0.7)
test_size =  len(x_data)-train_size
x_train = x_data[0:train_size,:]
x_test = x_data[train_size: len(x_data),:]
y_train = y_data[0:train_size]
y_test = x_data[train_size: len(x_data)]
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python

# Contruction de RNN
model = Sequential()

model.add(LSTM(units=50, return_sequences=True, input_shape=(x_train.shape[1],1)))

model.add(LSTM(units=50, return_sequences=True))

model.add(LSTM(units=50))
               

# Ajouter la couche de sortie
model.add(Dense(units=1))

# pr√©cuser la fonction cout / pour mettre √† jour les poids
model.compile(optimizer='adam', loss='mean_squared_error')

model.summary()
```

```python
# Entrainer le mod√®le
model.fit(x_train, y_train, epochs=100, batch_size=32)
```

```python
# Faire des  pr√©vison 
y_pred = model.predict(x_test)
```
